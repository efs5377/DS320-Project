```python
# Import packages
import pandas as pd
import country_converter as coco
from google.colab import files
```

```python
# Load in data
nuclear = pd.read_csv("/content/drive/MyDrive/DS320 Final Project/data/raw/Primary-energy-consumption-from-fossilfuels-nuclear-renewables.csv")
temp_country = pd.read_csv("/content/drive/MyDrive/DS320 Final Project/data/raw/GlobalLandTemperaturesByCountry.csv")
```

```python
# View data
nuclear.head(10) # Entity = Country
```

```python
# Rename columns
df_nuclear = nuclear.rename(columns = {"Entity":"Country"})
```

```python
# Removes any rows with NaN in the columns Country or Year
df_nuclear1 = df_nuclear.dropna(how='any', subset=['Country', 'Year'])
```

```python
df_nuclear1["short_name"] = coco.convert(names = df_nuclear1.Country.tolist(), to="name_short", not_found ="Unknown") 
```

```python
# Show continents
continents =  df_nuclear1[df_nuclear1["short_name"].str.contains("Unknown")]
continents = continents["Country"].unique()
print(continents)
```

```python
# View finalized dataframe
df_nuclear1.head(10)
```

```python
# View data
temp_country.head(10) # Make Country first column ... Make a df where dt is Year and it only has the years
```

```python
# Rearrange column order
df_temp_country = temp_country[['Country', 'dt', 'AverageTemperature', 'AverageTemperatureUncertainty']]
```

```python
# Removes any rows with NaN in the columns Country or dt
df_temp_country1 = df_temp_country.dropna(how='any', subset=['Country', 'dt'])
```

```python
df_temp_country1['DateTime'] = pd.DatetimeIndex(df_temp_country1["dt"]).year
```

```python
df_temp_country1
```

```python
#df_temp_country1["Year"] = df_temp_country1.dt.year["DateTime"]
df_temp_country2 =  df_temp_country1.groupby(['Country', 'DateTime'], as_index=False).mean()
```

```python
df_temp_country2
```

```python
df_temp_country2["short_name"] = coco.convert(names = df_temp_country2.Country.tolist(), to="name_short", not_found ="Unknown") 
```

```python
# Show continents
continents1 =  df_temp_country2[df_temp_country2["short_name"].str.contains("Unknown")]
continents1 = continents1["Country"].unique()
print(continents1)
```
